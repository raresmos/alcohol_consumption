# Distribution between Age and Daily Consumption (Dalc)

```python
corr_age_dalc = data['age'].corr(data['Dalc']).round()
corr_age_dalc
```

```python
age_distribution = data.groupby('Dalc')['age'].mean().round(2)
age_distribution
```

```python
plt.scatter(age_distribution.index , age_distribution.values)
plt.xlabel('Dalc')
plt.ylabel('Age')
```
